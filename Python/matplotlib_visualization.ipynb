```python
pivot1 = pd.pivot_table(df, values='eval', index='gender', columns='tenured_prof', aggfunc='mean')
print(pivot1)
```

```python
# Example 2: Sum of credits by division and English speaker status
pivot2 = pd.pivot_table(df, values='credits', index='division', columns='English_speaker', aggfunc='sum')
print(pivot2)
```

```python
pivot_data = df.pivot_table(index='gender',columns='division',values='age',aggfunc='count')
pivot_data.plot(kind='bar')
```

```python
# Median beauty score grouped by gender and minority
pivot_data = df.pivot_table(index='gender',values='beauty',columns='minority',aggfunc='median')
pivot_data.plot(kind='bar')
```
