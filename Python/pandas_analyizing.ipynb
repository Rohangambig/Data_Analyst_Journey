```python
# .describe()
import pandas as pd
file_path = 'C:/Users/rohan/OneDrive/Documents/notes/Desktop/Data analyst/Datasets/teachingratings.csv'

df = pd.read_csv(file_path)
df
```

```python
# find the minimum value of age 

min_index = df['age'].idxmin()
df.loc[min_index]
```

```python
# find the rows which has maximum age
max_age = df['age'].idxmax()
df.loc[max_age]
```

```python
# Finding the highest age of the man and women
df.groupby('gender')['age'].max()
```

```python
# Finding the least age of the man and women
df.groupby('gender')['age'].min()
```

```python
# finding the median of females and males based on age
df.groupby('gender')['age'].median()
```

```python
# finding the mean of females and males based on age
df.groupby('gender')['age'].mean()
```

```python
# use group by clause for age and gender
df.groupby(['gender','age'])[['gender','age']].size()
```

```python
# find is there age < 35 is present or not
new_df = df[ df['age'] == 35]
new_df
```
