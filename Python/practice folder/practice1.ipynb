```python
# Normal distribution probabilty distribution function
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4,4,1000)
y = norm.pdf(x,0,1)

plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
data1 = ['a','b','c','d']
data2 = [2,4,1,5]

plt.bar(data1,data2)
plt.grid(True)
plt.show()
```
