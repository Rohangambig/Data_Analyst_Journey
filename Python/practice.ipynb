## Pivot tables

```python
import pandas as pd 

cognizant_data =  pd.read_excel('C:/Users/rohan/OneDrive/Documents/notes/Desktop/Data analyst/Datasets/dataset Cognizant.xlsx')
cognizant_data.head(5)
```

```python
cognizant_data.drop(labels=['College Name','Status'],axis=1,inplace=True)
```

```python
# cognizant_data.drop(labels='Aptitude Assessment Date & Slot',axis=1,
# inplace=True)
cognizant_data.head(3)
```

```python
cognizant_data.info()
```

```python
pivot_data = cognizant_data.pivot_table(values='Phone Number',index='Program',aggfunc='mean',columns='Superset ID')
pivot_data
```

```python
# Use  group by class to calculate the number of students who enrolled for each course

programCount = cognizant_data.groupby('Course').size()
programCount
```

```python
# Ploting the line graph

import matplotlib.pyplot as plt
programCount.plot(kind='line')
plt.xticks(rotation=90,ha='right')
plt.xlabel('Courses')
plt.ylabel('Number of students')
plt.grid(True)
plt.show()
```

```python
# plotting the bar graph
import matplotlib.pyplot as plt
programCount.plot(kind='bar')
plt.xticks(rotation=90,ha='right')
plt.xlabel('Courses')
plt.ylabel('Number of students')
plt.grid(True)
plt.show()
```

```python
# plotting the histogram graph
import matplotlib.pyplot as plt
programCount.plot(kind='hist')
plt.xticks(rotation=90,ha='right')
plt.xlabel('Courses')
plt.ylabel('Number of students')
plt.grid(True)
plt.show()
```

```python
# plotting of group bar chart

program_count = cognizant_data.groupby(['Program','Course'])[['Course','Program']].value_counts()
program_count
```
