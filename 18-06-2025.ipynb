```python
df1 = pd.DataFrame(
{"a" : [4, 5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]},
index = [1, 2, 3])


df2 = pd.DataFrame(
{"a" : [4, 6],
"b" : [7, 9],
"c" : [10,  12]},
index = [1, 3])
```

```python
# append the row of the dataFrames

pd.concat([df1,df2],ignore_index=True)
```
